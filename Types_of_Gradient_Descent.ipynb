## **Gradient Descent Example**

Let‚Äôs minimize the function:

$$f(w)=w^2$$


$$\frac{dw}{df} = 2w$$



Using gradient descent to find
ùë§
w that minimizes
ùëì(ùë§)


**Python Code for GD**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def f(w):
    return w**2

def gradient(w):
    return 2 * w

# Initialize parameters
w = 5  # Start at w=5
learning_rate = 0.1
iterations = 20

# Store values for visualization
w_history = [w]

# Gradient Descent Loop
for i in range(iterations):
    grad = gradient(w)  # Compute gradient
    w = w - learning_rate * grad  # Update weight
    print("Gradient is - ", grad)

    print("Updated weight is ", w)
    w_history.append(w)

# Plot
plt.plot(range(len(w_history)), w_history, marker='o')
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.title("Gradient Descent Optimization")
plt.show()
```

# **Types of Gradient Descent**

*   **Batch Gradient Descent** (updates after computing gradient on the entire dataset).
*   **Stochastic Gradient Descent** (SGD) (updates after each data point).
*   **Mini-Batch Gradient Descent** (updates after computing on a small batch).








**Batch Gradient Descent**

"Batch Gradient Descent is an optimization algorithm used to minimize a loss function by updating model parameters using the entire dataset in each iteration."

**How It Works:**


*   Compute the gradient (derivative) of the loss function using all training examples.
*   Update the model parameters (e.g., weights) using the computed gradient.
*   Repeat until convergence (i.e., the loss stops decreasing significantly).

**Algorithm:**

1. Initialize weights ùë§ and bias ùëè.

2. For each iteration:
* Compute the gradient of the loss function using the entire training dataset.
* Update parameters: (components specific)

$$
w = w - \eta \frac{\partial L}{\partial w}
$$

$$
b = b - \eta \frac{\partial L}{\partial b}
$$


3. Repeat until convergence.

***General Batch Gradient Descent Formula (for all parameters Œ∏)***

$$
\theta = \theta - \alpha \cdot \frac{1}{m} \sum_{i=1}^{m} \nabla J(\theta, x^{(i)}, y^{(i)})
$$


* $Œ∏$ = model parameters (weights, biases, etc.)

* $ùõº$ = learning rate (step size)
* $ùëö$ = total number of training examples
* $(ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$ = training data points

* $‚àáùêΩ(\theta, ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$
 = gradient of the loss function with respect to ùúÉ


**Python Code for BGD**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def f(w):
    return w**2

def gradient(w):
    return 2 * w  # Derivative of w^2 is 2w

# Simulating a batch dataset (multiple values of w)
w_values = np.array([5, 4, 3, 2, 1])  # Example batch dataset
learning_rate = 0.1
iterations = 20

# Initialize w as the mean of the batch (starting point)
w = np.mean(w_values)
w_history = [w]  # Store weight updates

# Batch Gradient Descent Loop
for i in range(iterations):
    grad = np.mean(gradient(w_values))  # Compute gradient over batch
    w = w - learning_rate * grad  # Update weight
    w_history.append(w)

# Plot
plt.plot(range(len(w_history)), w_history, marker='o', label="Batch GD")
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.title("Batch Gradient Descent Optimization")
plt.legend()
plt.show()
```

**Advantages:**

* More stable updates (since it uses the entire dataset).

* Converges smoothly to the minimum.

**Limitations:**
* Slow for large datasets.
* Requires high memory since the entire dataset is loaded at once.

**Stochastic Gradient Descent**

"Stochastic Gradient Descent is an optimization algorithm that updates model parameters by computing the gradient using a single training example at each iteration."

**How It Works:**

* Select a single random training example from the dataset.
* Compute the gradient (derivative) of the loss function using this single example.
* Update the model parameters (e.g., weights) immediately using the computed gradient.
* Repeat for each training example, iterating multiple times over the dataset until convergence.

**Algorithm:**

1. Initialize weights ùë§ and bias ùëè.
For each iteration:

* Pick one random data point
$(ùë•_ùëñ,ùë¶_ùëñ)$

* Compute the gradient using this single sample.
Update parameters:

 $$ w = w - \eta \frac{\partial L}{\partial w} $$

 $$ b = b - \eta \frac{\partial L}{\partial b} $$

* Repeat for multiple iterations.


***General Stochastic Gradient Descent Formula***

$$
\theta = \theta - \alpha \cdot \nabla J(\theta, x^{(i)}, y^{(i)})
$$


* $Œ∏$ = model parameters (weights, biases, etc.)

* $ùõº$ = learning rate (step size)

* $(ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$ = A single randomly chosen training example

* $‚àáùêΩ(\theta, ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$
 = Gradient of the loss function computed using only this one training example

***Python Code***

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def f(w):
    return w**2

def gradient(w):
    return 2 * w  # Derivative of w^2 is 2w

# Simulating a dataset (random values for w)
np.random.seed(42)
w_values = np.random.uniform(-5, 5, size=10)  # 10 random samples
learning_rate = 0.1
iterations = 20

# Initialize weight randomly
w = np.random.choice(w_values)
w_history = [w]  # Store weight updates

# Stochastic Gradient Descent Loop
for i in range(iterations):
    random_index = np.random.randint(len(w_values))  # Pick a random sample
    w_sample = w_values[random_index]  # Get one random sample
    grad = gradient(w_sample)  # Compute gradient for this sample
    w = w - learning_rate * grad  # Update weight
    w_history.append(w)

# Plot
plt.plot(range(len(w_history)), w_history, marker='o', label="SGD")
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.title("Stochastic Gradient Descent Optimization")
plt.legend()
plt.show()
```

**Advantages:**

* Faster updates, making it useful for large datasets.

* Works well for online learning (real-time updates).

**Limitations:**
* Highly noisy updates, which may lead to oscillations instead of smooth convergence.
* Doesn't always converge to the global minimum but fluctuates around it.

**Mini-Batch Gradient Descent**

"Mini-Batch Gradient Descent is an optimization algorithm that updates model parameters by computing the gradient using a small subset (mini-batch) of the training data at each iteration."

**How It Works:**
1. Divide the dataset into small mini-batches of size ùêµ (e.g., 32, 64, 128).
2. For each mini-batch:
  * Compute the gradient (derivative) of the loss function using only the examples in this batch.
  * Update the model parameters (e.g., weights) using the computed gradient.
3. Repeat for all mini-batches in the dataset, iterating multiple times until convergence.

**Algorithm:**

1. Initialize weights ùë§ and bias ùëè.
For each iteration:
2. For each iteration:
  * Select a small batch of random samples from the dataset.
  * Compute the gradient using this batch.
  * Update parameters using the batch average gradient.
  * Repeat until convergence.

***General Mini-Batch Gradient Descent Formula (for all parameters Œ∏)***

$$
\theta = \theta - \alpha \cdot \frac{1}{B} \sum_{i=1}^{B} \nabla J(\theta, x^{(i)}, y^{(i)})
$$


* $Œ∏$ = model parameters (weights, biases, etc.)

* $ùõº$ = learning rate (step size)
* $B$ = Mini-batch size (a subset of the dataset, e.g., 32, 64)
* $(ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$ = Training examples in the current mini-batch

* $‚àáùêΩ(\theta, ùë•^{(ùëñ)},ùë¶^{(ùëñ)})$
 = Gradient of the loss function computed using only the mini-batch

***Python Code***

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and derivative
def f(w):
    return w**2

def gradient(w):
    return 2 * w  # Derivative of w^2 is 2w

# Simulating a dataset (random values for w)
np.random.seed(42)
w_values = np.random.uniform(-5, 5, size=10)  # 10 random samples
learning_rate = 0.1
iterations = 20
batch_size = 3  # Define mini-batch size

# Initialize weight randomly
w = np.mean(w_values[:batch_size])  # Start with the mean of the first mini-batch
w_history = [w]  # Store weight updates

# Mini-Batch Gradient Descent Loop
for i in range(iterations):
    batch_indices = np.random.choice(len(w_values), batch_size, replace=False)  # Random mini-batch
    w_batch = w_values[batch_indices]  # Select batch samples
    grad = np.mean(gradient(w_batch))  # Compute gradient over the mini-batch
    w = w - learning_rate * grad  # Update weight
    w_history.append(w)

# Plot
plt.plot(range(len(w_history)), w_history, marker='o', label="Mini-Batch GD")
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.title("Mini-Batch Gradient Descent Optimization")
plt.legend()
plt.show()
```

**Advantages:**

* Balances stability and speed (less noisy than SGD, faster than BGD).

* Works well with hardware acceleration (GPU training).

**Limitations:**
* Needs tuning (batch size affects performance).
* Not as smooth as BGD and not as fast as SGD.

# **Challenges in Gradient Descent**

**1. Choosing the right learning rate:**

  * Too high ‚Üí Divergence.
  * Too low ‚Üí Slow convergence.

**2. Local Minima & Saddle Points:**

  * SGD helps escape local minima due to randomness.

**3. Vanishing & Exploding Gradients:**

  * Use batch normalization or gradient clipping.

**4. Convergence Issues:**

  * Adaptive optimizers like Adam can help.
