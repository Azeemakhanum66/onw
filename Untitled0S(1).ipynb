```python
train_datagen=ImageDataGenerator(rescale=1./255,horizontal_flip=True)
validation_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_dir='/content/Output/train'
train_gen=train_datagen.flow_from_directory(train_dir,target_size=(224,224),batch_size=10,class_mode='categorical')
validation_dir='/content/Output/val'
validation_gen=validation_datagen.flow_from_directory(validation_dir,target_size=(224,224),batch_size=10,class_mode='categorical')
test_dir='/content/Output/test'
test_gen=test_datagen.flow_from_directory(test_dir,target_size=(224,224),batch_size=10,class_mode='categorical')
```

```python
pre_train_model=tf.keras.applications.ConvNeXtBase(input_shape=(224,224,3),include_top=False,weights='imagenet')

for layer in pre_train_model.layers:
  layer.trainable=False
```

```python
pre_train_model.summary()
```

```python
last_layer=pre_train_model.get_layer('layer_normalization_1')
last_output=last_layer.output
print(last_output)
```

```python
x=Flatten()(last_output)
x=Dense(1024,activation='relu')(x)
x=Dense(1024,activation='relu')(x)
x=Dense(1024,activation='relu')(x)
x=Dense(5,activation='softmax')(x)
```

```python
metrics=['accuracy',tf.keras.metrics.Precision(),tf.keras.metrics.Recall()]
model=Model(pre_train_model.input,x)
model.compile(optimizer='RMSprop',loss='categorical_crossentropy',metrics=metrics)
```

```python
model.summary()
```

```python
history = model.fit(train_gen, epochs=2, validation_data=validation_gen)
```

```python
def calculate_test_preds(source, model):
    predictions = []
    for imgname in os.listdir(source):
        img_path = os.path.join(source, imgname)
        print(f"Processing image: {imgname}")  # Debugging output
        try:
            img = image.load_img(img_path, target_size=(224, 224))
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            preds = model.predict(img_array)
            predictions.append(preds)
        except Exception as e:
            print(f"Error loading image {imgname}: {e}")
    print(f"Total predictions made: {len(predictions)}")  # Debugging output
    return predictions
```

```python
predictions = calculate_test_preds(source='/content/test',model=model)
```

```python
true_labels = []
for filename in os.listdir('/content/test'):
  for cat in os.listdir('/content/Data'):
    category = os.path.join('/content/Data',cat)
    if filename in os.listdir(category):
      if cat == 'AcneandRosaceaPhotos':
        true_labels.append(0)
      if cat == 'ActinicKeratosisBasalCell ':
        true_labels.append(1)
      if cat == 'AtopicDermatitisPhotos':
        true_labels.append(2)
      if cat == 'BullousDiseasePhotos':
        true_labels.append(3)
      if cat == 'CellulitisImpetigo':
        true_labels.append(4)
```

```python
import numpy as np
from sklearn.metrics import classification_report

predicted_classes = np.argmax(predictions, axis=2).flatten()

print(classification_report(true_labels, predicted_classes))
```

```python
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Convert predictions to 1D array of predicted class labels
predicted_classes = np.argmax(predictions, axis=2).flatten()

# Now you can compute the confusion matrix
cm = confusion_matrix(true_labels, predicted_classes)
print(cm)
```

```python
import seaborn
import numpy as np
seaborn.set(font_scale=2)
fig, ax = plt.subplots(figsize=(12,10))
classes = ['AcneandRosaceaPhotos','ActinicKeratosisBasalCell','AtopicDermatitisPhotos','BullousDiseasePhotos','CellulitisImpetigo']
seaborn.heatmap(cm, annot=True, fmt='d',xticklabels=classes, yticklabels=classes,cmap='YlGnBu')
bottom, top = ax.get_ylim()
'''ax.set_ylim(bottom + 0.5, top - 0.5)'''
plt.tight_layout()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
